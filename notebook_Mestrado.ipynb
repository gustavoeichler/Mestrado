```python
SNRNU_q = list()
SNRFU_q = list()
R1NU_q = list()
R2FU_q = list()
for qos in range(1,10):
    SNRNU_d = list()
    SNRFU_d = list()
    R1NU_d= list()
    R2FU_d=list()
    for d1 in range(10,120):
        
        R1NU= list()
        R2FU=list()
        SNRNU = list()
        SNRFU = list()
        for d2 in range(d1,121):
            R1 = list()
            R2 = list()
            SNR1 = list()
            SNR2 = list()    
            for i in range(120):
                NU = OM.Canal(d1,2) 
                FU = OM.Canal(d2,2)
                Saida = OM.Otimizar_NOMA(NU,FU,No,P,1,qos,1)
                R1.append(Saida[1])
                R2.append(Saida[2])
                SNR1.append(Saida[4])
                SNR2.append(Saida[5])
            R1NU.append(np.mean(R1))
            R2FU.append(np.mean(R2))
            SNRNU.append(np.mean(SNR1))
            SNRFU.append(np.mean(SNR2))
        R1NU_d.append(np.mean(R1NU))
        R2FU_d.append(np.mean(R2FU))
        SNRNU_d.append(np.mean(SNRNU))
        SNRFU_d.append(np.mean(SNRFU))    
    R1NU_q.append(R1NU_d)
    R2FU_q.append(R2FU_d)
    SNRNU_q.append(SNRNU_d)
    SNRFU_q.append(SNRFU_d)


    
            
```
