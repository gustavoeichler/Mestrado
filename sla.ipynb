```python
from ortools.linear_solver import pywraplp
# Create the linear solver with the GLOP backend.
solver = pywraplp.Solver.CreateSolver('GLOP')

custos = sub_util
x={}
for i in range(len(custos.columns)):
    for j in range(len(custos.index)):
        x[j, i] = solver.IntVar(0, 1, f'NU:{custos.index[j]}_FU:{custos.columns[i]}')

for i in range(len(custos.columns)):
    for j in range(len(custos.index)):
        solver.Add(([x[i, j]] == 10))

solver.NumConstraints()
```

```python
# Implemente uma solucao para o problema:
# max sum xij Aij
# s.t.
#   sum_i xij = 1  ... para todo j
#   sum_j xij = 1  ... para todo i
# 	0 <= xij <= 1

#Vamos importar o solver linear do ORTOOLS
from __future__ import print_function
from ortools.linear_solver import pywraplp
#Vamos importar o numpy como uma ferramenta auxiliar 
import numpy as np
import math
#Vamos declarar o solver considerando o Google's Linear Optimization Programming system (GLOP)
solver = pywraplp.Solver('simple_lp_program', pywraplp.Solver.GLOP_LINEAR_PROGRAMMING)


def mi2ai(i,j,line_length):
	return i*N+j

def ai2mi(index, line_length):
	return [math.floor(index/line_length), index%line_length]


# [Definicao dos parametros]
#Numero de usuários e espaços
N = len(custos.columns)
# Matriz de interesses
A = custos
# [Fim da definicao dos paramentros]

# [Definicao das variaveis de decisao]
#Vamos criar as variaveis numericas do problema de otimizacao
#Formato: solver.NumVar(limite inferior, limite superior, rotulo)
# x = solver.NumVar(0, 1, 'x')
# y = solver.NumVar(0, 1, 'y')
# x = solver.NewIntVar(0, 1, 'x')
x = []
# i*j variaveis serao necessarias
for i in range(1,N+1):
	for j in range(1,N+1):
		x.append(solver.NumVar(0, 1, str(i)+","+str(j)))
print ('Numero de variaveis =', solver.NumVariables())
# [Fim da definicao das variaveis de decisao]

# [Definicao das restricoes]
# Vamos criar as restricoes do problema para as variaveis de decisao
# cts = []
# Caso saiba de antemão o numero de restricoes - no nosso caso 6
# cts = [[]] * 6
# Vamos criar um indice para as restricoes
head = 0
# Exemplo: -infinito <= x + y <= 2
# ######## ct = solver.Constraint(-solver.infinity(),2,'ct')
# ######## ct.SetCoefficient(x, 1)
# ######## ct.SetCoefficient(y, 1)
# Restricao Todo presente so pode pertencer a uma pessoa
# Restricao linear: sum_i xij = 1; para todo j em 1,2,3 ..., N
#### Definir o laço do somatorio
#for j in range(0,N):
	#Vamos criar a restricao para cada presente 
#	ct = solver.Constraint(0,1, str(head))
#	for i in range(0,N):
#		ct.SetCoefficient(x[mi2ai(i,j,N)], 1)
#	head = head+1
# Restricao Toda pessoa so pode pegar um presente
# Restricao linear: sum_j xij = 1; para todo i em 1,2,3 ..., N
# Definir o laço do somatorio
#for i in range(0,N):
	#Vamos criar a restricao para cada presente 
#	ct=solver.Constraint(0,1, str(head))
#	for j in range(i,N):
##		ct.SetCoefficient(x[mi2ai(i,j,N)], 1)
#	head = head+1
####ct = solver.Constraint(0,N/2, str(head))
#for i in range(0,N):
#    ct.SetCoefficient(x[mi2ai(i,i,N)], 1)
head = head+1
	
for i in range(N):
	head = head+1
	ct = solver.Constraint(1,1, str(head))
	for j in range(N):
		ct.SetCoefficient(x[mi2ai(i,j,N)], 1)
		ct.SetCoefficient(x[mi2ai(j,i,N)], 1)
			
head = head+1

#ct = solver.Constraint(N/2,N, str(head))
#for i in range(0,N):
#	for j in range(0,N):
#		ct.SetCoefficient(x[mi2ai(i,j,N)], 1)

#for i in range(0,N):	
#    for j in range(i,N):
#        ct.SetCoefficient(x[mi2ai(i,j,N)], 1)
# [Fim da definicao das restricoes]

#[Funcao objetivo]
# Definiremos a funcao objetivo 
objective = solver.Objective()
# Maximizar o interesse de todos
# Objetivo linear: max sum_ij xij*Aij 
# Agora, precisaremos colocar os coeficientes multiplicativos de xij
# Exemplo: objective.SetCoefficient(y, 1)
#Criar o laco para variar entre todas as variaveis i,j em NxN
for i in range(N):
    for j in range(N):
        objective.SetCoefficient(x[mi2ai(i,j,N)], A.iloc[i,j])      
        # Definiremos o problema como uma maximizacao
objective.SetMaximization()

# #Executaremos o solver
solver.Solve()

#Imprimiremos a solucao
print('Solucao:')
print('Valor objetivo =', objective.Value())
strmatrix=''
for i in range(0,N):
	strmatrix+='[ '
	for j in range(0,N):

		strmatrix += str(int(x[mi2ai(i,j,N)].solution_value()))+" "
	strmatrix+=']\n'
print(strmatrix)
strmatrix=''
pair = 0
usuario_atendido = list(DF_NOMA.columns)
for i in range(0,N):

	for j in range(0,N):

		if int(x[mi2ai(i,j,N)].solution_value()) ==1:
			if i == j:
				print(f'OMA User: {DF_NOMA.columns[j]}')
				print(f'Usou {A.iloc[i,j]*100}% do suportado') 
				usuario_atendido.remove(DF_NOMA.columns[j])
			else:
				
				print(f'NOMA Pair {pair} Near  User: {DF_NOMA.columns[j]}') 
				print(f'NOMA Pair {pair} Far  User: {DF_NOMA.index[i]}')
				print(f'Usaram {A.iloc[i,j]*100}% do suportado') 
				usuario_atendido.remove(DF_NOMA.columns[j])
				usuario_atendido.remove(DF_NOMA.index[i])
				pair = pair +1 

usuario_atendido
```
